```python
#changing the 'date' column datatype into datetime64
data['date']=pd.to_datetime(data['date'])
```

```python
#these columns contains string values which needs to be removed (landed 'of' attempted)
Columns = ['R_SIG_STR.', 'B_SIG_STR.', 'R_TOTAL_STR.', 'B_TOTAL_STR.',
       'R_TD', 'B_TD', 'R_HEAD', 'B_HEAD', 'R_BODY','B_BODY', 'R_LEG', 'B_LEG', 
        'R_DISTANCE', 'B_DISTANCE', 'R_CLINCH','B_CLINCH', 'R_GROUND', 'B_GROUND']
```

```python
#splitting above columns into attempted and landed
for col in Columns:
    data[col+'_att'] = data[col].apply(lambda X: int(X.split('of')[1]))
    data[col+'_land'] = data[col].apply(lambda X: int(X.split('of')[0]))
    
data.drop(Columns, axis=1, inplace=True)#trim
```

```python
#these columns contains percentage sign, which needs to be removed
Col = ['R_SIG_STR_pct','B_SIG_STR_pct','R_TD_pct','B_TD_pct']
```

```python
#removng '%' sign and changing its datatype to float
for c in Col:
    data[c] = data[c].apply(lambda X: float(X.replace('%', ''))/100)
```

```python
#creating function to get weight class
def Division(X):
    for Division in weight_classes:
        if Division in X:
            return Division
    if X == 'Catch Weight Bout' or 'Catchweight Bout':
        return 'Catch Weight'
    else:
        return 'Open Weight'
```

```python
weight_classes = ['Women\'s Strawweight', 'Women\'s Bantamweight', 
                  'Women\'s Featherweight', 'Women\'s Flyweight', 'Lightweight', 
                  'Welterweight', 'Middleweight','Light Heavyweight', 
                  'Heavyweight', 'Featherweight','Bantamweight', 'Flyweight', 'Open Weight']

data['weight_class'] = data['Fight_type'].apply(Division)
```

As shown above, all the features are categorical and Machine Learning algorithms takes only numeric input.
So we need to convert these categorical variables to numeric and drop unnecessary features.

Categorical variables are of two types:
1. Nominal - they are in the form of names,labels etc which cannot be ordered
   eg. (win_by)
2. Ordinal - they can be ordered
   eg. (weight_class) 

Ordinal Categories

```python
#changing the datatype of target column
data['Winner'] = data['Winner'].astype('category')
data['weight_class'] = data['weight_class'].astype('category')
data['Type of Fight'] = data['Type of Fight'].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#name of columns which are in ordered form
ordinal = data[['weight_class','Winner','Type of Fight','Gender']]
#label encoding on ordinal columns
for col in ordinal.columns:
    ordinal[col] = le.fit_transform(ordinal[col])
```

```python
#checking the top 5 rows of ordinal data after label encoding
ordinal.head()
```

Nominal categories

```python
#converting categorical variable to dummy
nominal = pd.get_dummies(data,columns=['win_by'],drop_first=True)
```

```python
nominal
```

```python
#dropping the unnecessary columns
dummy = nominal.drop(['R_fighter',
 'B_fighter',
 'last_round_time',
 'Format',
 'Referee',
 'location','date',
 'Fight_type',
 'weight_class',
 'Country',
 'Type of Fight','Gender','Winner'],axis=1)
```

```python
#concatenating final data
cleaned_data = pd.concat([dummy,ordinal],axis=1)
```

```python
#checking top 5 rows of final data
cleaned_data.head()
```

```python
#Predictor variables (independent variables)
x = cleaned_data.drop('Winner',axis=1).values
```

```python
#Target variable (dependent variable)
y = cleaned_data[['Winner']].values
```

## Which model to select?

```python
#Logistic Regression using cross validation
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
cross_val_score(logreg,x,y,cv=10,scoring='accuracy').mean()
```

```python
#KNN using cross validation
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
cross_val_score(knn,x,y,cv=10,scoring='accuracy').mean()
```

```python
#Random Forest using cross validation
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
cross_val_score(rf,x,y,cv=10,scoring='accuracy').mean()
```

Best score was given by Logistic Regression model using Cross Validation

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=33) 
from sklearn.linear_model import LogisticRegression   #importing Logistic Regression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
ac = accuracy_score(y_test,y_pred)
print('ACCURACY SCORE:{}'.format(ac))
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
```

### Hyperparameter Tuning

```python
lr.get_params
```

```python
param_grid = {"penalty": ["l2"], "fit_intercept": [True, False], 
              "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"],
              "max_iter": [50, 100, 200], "warm_start": [True, False]}
```

```python
from sklearn.model_selection import GridSearchCV
gcv = GridSearchCV(lr,param_grid=param_grid,cv=10)
gcv.fit(x,y)
```

```python
print('BEST PARAMETERS : {}'.format(gcv.best_params_))
print('BEST SCORE : {}'.format(gcv.best_score_))
```

After tuning the parameters the overall accuracy of model was increased.
